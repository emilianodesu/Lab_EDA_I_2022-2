Gráfica 3

```python
# arange recibe el parámetro inicial el parámetro final
# y el intervalo de incremento
t = np.arange(0., 5., 0.2)

plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
# plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')

plt.show()
```

Gráfica 4

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

# Primera figura
plt.figure(1)

# Primera gráfica
plt.subplot(121)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

# Segunda gráfica
plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.show()
```

Gráfica 5

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2,1), xytext=(3, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.ylim(-2,2)
plt.show()
```

Gráfica 6

```python
country = ("Alemania", "España", "Francia", "Portugal")
position_y = np.arange(len(country))
population = (82.67, 45.56, 66.9, 10.32)

plt.barh(position_y, population, align="center")

plt.yticks(position_y, country)
plt.xlabel('Población (millones)')
plt.title("Población por país")

plt.show()
```

Gráfica 7

```python
data = [[72.81, 46.81, 52.4, 11.49], [79.43, 58.51, 57.25, 14.95], [82.67, 66.9, 65.64, 16.62]]
axisX = np.arange(4)
plt.bar(axisX + 0.0, data[0], color="b", width=0.25)
plt.bar(axisX + 0.3, data[1], color="c", width=0.25)
plt.bar(axisX + 0.6, data[2], color="m", width=0.25)
plt.xticks(axisX+0.25, ["Alemania", "Francia", "Reino Unido", "Países Bajos"])

plt.show()
```
